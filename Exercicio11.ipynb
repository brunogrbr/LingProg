```python
def eager_fibonacci (n):
    l = [1, 1]
    for i in range (n - 2):
        l.append(sum(l[-2:]))
    return l

print (eager_fibonacci(5))
```

```python
def lazy_fibonacci ():
    yield 1
    yield 1
    l = [1, 1]
    while True:
        l = [l[-1], sum(l[-2:])]
        yield l[-1]
        
g = lazy_fibonacci ()
print (next(g))
print (next(g))
print (next(g))
print (next(g))
print (next(g))
```

```python
def fibonacci ():
    yield 1
    yield 1
    l = [1, 1]
    while True:
        l = [l[-1], sum(l[-2:])]
        yield l[-1]
        
def tribonacci ():
    yield 0
    yield 1
    yield 1
    l = [0, 1, 1]
    while True:
        l = [l[-2], l[-1], sum(l[-3:])]
        yield l[-1]

for i, (f, t) in enumerate (zip (fibonacci(), tribonacci())):
    if i == 10:
        break
    print (f, t)
```

```python
import random

SENTENCES = [
    'How are you?',
    'Fine, thank you',
    'Nothing much',
    'Just chillin',
    'Bye'
]

def speaker():
    while True:
        yield random.choice(SENTENCES)
        
def replier():
    while True:
        recv = yield
        print("Received: %s"%recv)
        if recv == "Bye":
            break
        print("Replied: %s" % random.choice(SENTENCES))
        
s = speaker()
r = replier()

#inicia o replier
r.send(None)
while True:
    #é o mesmo que chamar o next
    recv = s.send(None)
    try:
        #enviamos o que o speaker gerou para o replier
        r.send(recv)
    except StopIteration:
        break
```
