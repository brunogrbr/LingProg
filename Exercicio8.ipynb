1. Escreva uma função que calcula a soma dos n primeiros números naturais, dado que n é recebido como parâmetro. Escreva um decorator que garanta que o valor recebido é natural (maior ou igual a 1)

```python
def somaDecorator(fnc, num):
    
    if num >= 1:
    
        def decorator():
        
            fnc(num)
        
        return decorator
    
    return lambda: print("Numero invalido %i" % num)

def soma(numero):
    
    soma = 0
    
    for n in range(1, numero + 1):
     
        soma += n
        
    print(soma)
```

```python
def somaDecorator(fnc):
            
    def decorator(*args):
    
        if args[0] >= 1:
        
            fnc(*args)
            
        else:
            
            print("Numero invalido %i " % args[0])
        
    return decorator

@somaDecorator
def soma(numero):
    
    soma = 0
    
    for n in range(1, numero + 1):
     
        soma += n
        
    print(soma)
```

```python
class SomaDecorator:
    
    def __init__(self, fnc):
        
        self.fnc = fnc
        
    def __call__(self, num):
        
        self.fnc(num)
        
@SomaDecorator
def soma(num):
    
    soma = 0
    
    for n in range(1, num + 1):
     
        soma += n
        
    print(soma)
```

Escreva uma função que recebe uma lista de triplas e, para cada uma, gera uma equação do segundo grau considerando que os elementos da tripla são os coeficientes usualmente denominados a, b e c da equação. Note que a sua função deverá devolver uma lista de equações. A geração das equações deve ser feita por meio de, evidentemente, decorators.

```python
def equacaoDecorator(**kwargs):
     
    def decorator():
    
        return kwargs.get('fnc')(kwargs.get('lista'))
        
    return decorator
        
def equacao(lista):
    
    for l in lista:
        
        print("\n%i * x^2 + %i * x + %i " % (l[0], l[1], l[2]))
```

```python
def equacaoDecorator(fnc):
     
    def decorator(*args):
        
        return fnc(*args)
        
    return decorator

@equacaoDecorator       
def equacao(lista):
    
    for l in lista:
        
        print("\n%i * x^2 + %i * x + %i " % (l[0], l[1], l[2]))
```

```python
class EquacaoDecorator:
    
    def __init__(self, fnc):
        
        self.fnc = fnc
        
    def __call__(self, lista):
        
        self.fnc(lista)

@EquacaoDecorator       
def equacao(lista):
    
    for l in lista:
        
        print("\n%i * x^2 + %i * x + %i " % (l[0], l[1], l[2])
```

3. Escreva uma função que exibe uma lista recebida como parâmetro. Ela deve, contudo, ordenar a lista antes. A ordenação deve ser feita por meio de um decorator.

```python
def ordernaLista(fnc, lista):
    
    def decorator():
            
        fnc(lista)
        
    return decorator
        
def listar(lista):
    
    lista.sort()
    
    print(lista)
        
```

```python
def ordernaLista(fnc):
    
    def decorator(*args):
            
        fnc(*args)
        
    return decorator

@ordernaLista        
def listar(lista):
    
    lista.sort()
    
    print(lista)
```

```python
class OrdernaLista:
    
    def __init__(self, fnc):
        
        self.fnc = fnc
        
    def __call__(self, lista):
        
        self.fnc(lista)

@OrdernaLista        
def listar(lista):
    
    lista.sort()
    
    print(lista)
```

4. Escreva uma classe para representar um café que, a princípio, tem somente preço.Um cafezinho custa 5 reais. Escreva classes para representar os adicionais: palitos dechocolate (0,50 cents), espuma de leite (0,20 cents), caramelo (0,10 cents) e canela (0,30 cents). Crie um objeto café e, a seguir, faça um menuzinho em que o usuário pode ficar indefinidamente escolhendo adicionais: 1 para chocolate, 2 para espuma de leite, 3 para caramelo e 4 para canela. A cada adicional escolhido, decore o objetocafé. No final, mostre o preço total.

```python
class Cafe:
    
    def __init__(self):
        
        self._preco = 5
        self._total = self._preco
        
    def __call__(self, obj):
        
        self._total += obj.getValor()
        
        return obj
    
    def getTotal(self):
        
        return self._total
       
    
class PalitoChocolate:
    
    def __init__(self):
        
        self._preco = 0.5
       
    def getValor(self):
        
        return self._preco
    
class EspumaLeite:
    
    def __init__(self):
        
        self._preco = 0.2
        
    def getValor(self):
        
        return self._preco
    
class Caramelo:
    
    def __init__(self):
        
        self._preco = 0.10
        
    def getValor(self):
        
        return self._preco
    
class Canela:
    
    def __init__(self):
        
        self._preco = 0.30
        
    def getValor(self):
        
        return self._preco
    
    
cafe = Cafe()

while True:
     
    print('\n--Menu--\n[1] Chocolate\n[2] Espuma de leite\n[3] Caramelo\n[4] Canela\n[0] Sair\n')
    
    opcao = int(input('Digite uma opcao para adicionar: '))
    
    if(opcao == 1):
    
        cafe(PalitoChocolate())
        
    if(opcao == 2):
    
        cafe(EspumaLeite())    
    
    if(opcao == 3):
    
        cafe(Caramelo())
        
    if(opcao == 4):
    
        cafe(Canela())
        
    if(opcao == 0):
        
        break;
        
    print('\n--> Total Cafe: R$ %.2f \n' % cafe.getTotal())
```
